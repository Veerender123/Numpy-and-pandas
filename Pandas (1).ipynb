# SERIES

```python
import pandas as pd
import numpy as np
```

# DEFAULT INDEXING

```python
k=np.array([1,2,3,4,'a'])
s=pd.Series(k);
s
```

# CUSTOM INDEXING

```python
t=np.array([1,2,3,4,5])
s=pd.Series(t,index=['a','b','c','d','e']);
s
```

# SERIES FROM DICTONARY,TUPLE,SET

```python
e=(1,2,3,4,5)
g={'name':['veerender','askhaya'],
   'fname':'ASHOK',
   'mname':'kavitha'
}
#set is unordered so we cant make it into series
t=pd.Series(e)
d=pd.Series(g)
print(t)
print(d)
```

# DATA ACCESSING

```python
t=np.array([1,2,3,4,5])
s=pd.Series(t,index=['a','b','c','d','e']);
print('indexing',s[0])
print(s['a'],"custom indexing")
print(s['b':])#sclicing using custom index
print(s[['a','b','e']])#multiple accessing
```
